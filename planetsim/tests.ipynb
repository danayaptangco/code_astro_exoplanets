```python
from run_sim import Planet, Star, Transit
import pytest
```

```python
def test_planet_conversion():
    planet = Planet(radius=1, orbital_period=1)

    expected_radius_m = 7.149e+7  
    expected_orbital_period_s = 86400


    assert planet.radius == pytest.approx(expected_radius_m)
    assert planet.orbital_period == expected_orbital_period_s

def test_star_conversion():
    star = Star(radius=1, mass=1)

    expected_radius_m = 6.957e+8 
    expected_mass_kg = 1.989e+30  

    assert star.radius == pytest.approx(expected_radius_m)
    assert star.mass == pytest.approx(expected_mass_kg)

test_planet_conversion()
test_star_conversion()
```
